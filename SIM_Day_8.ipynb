# an animated 3D plot where a sphere moves in a sinusoidal motion along the x, y, and z axes

 **y = np.sin(t + i * 0.1):** The y coordinate of the sphere oscillates in a sinusoidal pattern over time, but with a phase shift determined by i * 0.1. The value of i changes with each frame, making the sine wave move or "shift" over time.<br>
 **x = np.cos(t + i * 0.1):** Similarly, the x coordinate follows a cosine wave, but with the same time-dependent shift, creating a smooth, circular-like motion in the x-y plane.<br>
**z = t:** The z coordinate simply increases linearly with t, so the sphere moves upwards along the z-axis without oscillation.<br>
the sphere's motion in the x-y plane traces out a circular path, and as time progresses (due to the i factor), the sphere moves smoothly along the z-axis in a spiral-like pattern.

```python

# Create figure and axis
fig = plt.figure() #creates a figure object
ax = fig.add_subplot(111, projection='3d') # adds a 3D subplot to the figure. 111 refers to the grid layout for subplots (1 row, 1 column, 1st subplot),
#and projection='3d' specifies that this subplot will be 3D.

# Generate data for a moving sphere
def update_plot(i): #function to update the plot during each frame of the animation.
    ax.clear() #reset the plot before drawing a new frame
    t = np.linspace(0, 2 * np.pi, 100) #100 points between 0 and 2π, create the coordinates of the moving sphere, one complete revolution of circle
    x = np.cos(t + i * 0.1) #x coordinates of the moving sphere are calculated as the sine of t with a phase shift dependent on i
    y = np.sin(t + i * 0.1) #y coordinates of the moving sphere are calculated as the sine of t with a phase shift dependent on i
    z = t # linear motion of the sphere along the z-axis.
    ax.plot(x, y, z, label="3D Animation")
    ax.set_title("Simple 3D Animation Simulation")
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')

# Create an animation
ani = animation.FuncAnimation(fig, update_plot, frames=100, interval=100)
'''
fig: The figure where the plot is being drawn.
update_plot: The function that updates the plot on each frame.
frames=100: The number of frames in the animation (100 frames).
interval=100: The interval between each frame in milliseconds (100ms).'''

# Save animation as a video
ani.save('/content/3d_animation.mp4', writer='ffmpeg', fps=30)

# Display the video in the notebook
Video('/content/3d_animation.mp4')
```

# animation of a moving sphere in a circular motion along the XY plane in 3D space

```python
# Create figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the function for updating the plot with circular motion
def update_plot(i):
    ax.clear()  # Clear the previous plot

#set the path******
    # Set the range for the angle and define the circle
    t = np.linspace(0, 2 * np.pi, 100) #one complete revolution

    # Circular motion equation (parametric equations for a circle in 3D)
    radius = 5
    x = radius * np.cos(t + i * 0.1)  # Circular motion along X
    y = radius * np.sin(t + i * 0.1)  # Circular motion along Y
    z = np.zeros_like(t)  # Keep Z constant to move on the XY plane

    ax.plot(x, y, z, label="Moving Sphere", color='b')  # Plot the moving sphere path
    ax.scatter(x[-1], y[-1], z[-1], color='r', s=100)  # Add the sphere at the current position (red dot)  coordinates of the last position in the animation
#****************************

    # Set plot labels and title
    ax.set_title("Circular 3D Animation Simulation")
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')

    ax.set_xlim(-radius, radius)
    ax.set_ylim(-radius, radius)
    ax.set_zlim(-radius, radius)

# Create the animation
ani = animation.FuncAnimation(fig, update_plot, frames=100, interval=100)
#repeatedly calling update_plot(i) for 100 frames.
#100 ms between frames

# Save the animation as a video
ani.save('/content/3d_circular_motion.mp4', writer='ffmpeg', fps=30) #30 frames per second

# Display the video in the notebook
Video('/content/3d_circular_motion.mp4')
```
