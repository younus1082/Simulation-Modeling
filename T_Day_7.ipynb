# Chi Square Fit Test
Chi-Square Goodness-of-Fit Test in Python

## Import Necessary Libraries

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
````

## Step 1: Observed data (frequencies in each category)

````python
observed = np.array([47, 49, 40])  # Example: Replace with your data
````

## Step 2: Expected frequencies (assuming uniform distribution)

````python
total = np.sum(observed)
expected = np.full_like(observed, total / len(observed))  # Uniform distribution
````

## Step 3: Chi-square statistic calculation

````python
chi_square_stat = np.sum(((observed - expected) ** 2) / expected)
````

## Step 4: Degrees of freedom

````python
df = len(observed) - 1
````

## Step 5: Calculate the p-value and critical value

````python
p_value = 1 - stats.chi2.cdf(chi_square_stat, df)  # p-value
alpha = 0.05  # Significance level
critical_value = stats.chi2.ppf(1 - alpha, df)  # Critical value
````

## Display the results

````python
print(f'Chi-square statistic: {chi_square_stat:.2f}')
print(f'Degrees of freedom: {df}')
print(f'Critical value (0.05 significance level): {critical_value:.2f}')
print(f'P-value: {p_value:.4f}')
````

## Conclusion

````python
if chi_square_stat > critical_value:
    print('Reject the null hypothesis. Observed data does not match expected distribution.')
else:
    print('Fail to reject the null hypothesis. Observed data matches the expected distribution.')
````

## Plotting observed vs expected frequencies

````python
plt.bar(range(1, len(observed) + 1), observed, width=0.4, label='Observed', alpha=0.7)
plt.bar(np.arange(1, len(observed) + 1) + 0.4, expected, width=0.4, label='Expected', alpha=0.7) #You can’t do range(...) + 0.4 because range objects don’t support adding a float to every element directly.
plt.xlabel('Categories')
plt.ylabel('Frequencies')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.show()
````

#### Theory
The Chi-square goodness-of-fit test compares the observed frequencies to the expected frequencies and determines whether there is a significant difference between them. The test statistic is calculated as:
\[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
\]
Where:
- \(O_i\) = observed frequency in category \(i\)
- \(E_i\) = expected frequency in category \(i\)

Degrees of freedom (df) is calculated as:
\[
df = \text{{number of categories}} - 1
\]
The result is compared with the critical value from the Chi-square distribution at a specified significance level (\(\alpha = 0.05\)).

#### Steps
1. Collect the observed data (frequencies).
2. Define the expected distribution (e.g., uniform, normal).
3. Compute the expected frequencies based on the distribution.
4. Calculate the Chi-square statistic.
5. Compare the statistic with the Chi-square distribution critical value to accept or reject the null hypothesis.


### Explanation of the Code
1. **Observed Data**: The observed frequencies are defined in the `observed` array. This can be replaced with your own dataset.
2. **Expected Frequencies**: The expected frequencies are calculated assuming a uniform distribution. For a uniform distribution, each category is expected to have an equal proportion of the total sum of the observed data.
3. **Chi-Square Statistic**: We use the `chisquare` function from `scipy.stats` to calculate the Chi-square statistic and the p-value.
4. **Degrees of Freedom**: The degrees of freedom is calculated as `df = len(observed) - 1`.
5. **Critical Value**: The critical value at the 0.05 significance level is obtained from the Chi-square distribution using `stats.chi2.ppf`.
6. **Conclusion**: We compare the Chi-square statistic with the critical value to determine whether to reject the null hypothesis.
7. **Visualization**: A bar chart is plotted to compare the observed and expected frequencies for each category.

### Discussion and Conclusion
- **Chi-square statistic**: It quantifies the deviation between the observed and expected frequencies.
- **Degrees of Freedom (df)**: It represents the number of categories minus one. It affects the Chi-square distribution.
- **P-value**: The p-value indicates the probability of observing a statistic as extreme as the one calculated, given that the null hypothesis is true.
  - If the p-value is less than 0.05, reject the null hypothesis, indicating that the observed data does not fit the expected distribution.
  - If the p-value is greater than 0.05, fail to reject the null hypothesis, indicating that the observed data reasonably fits the expected distribution.
