# Uniform Distribution with Random Numbers

A uniform distribution is a type of probability distribution where all outcomes are equally likely. In other words, every event in a given range has the same chance of happening. For example, when you roll a fair six-sided die, the chances of rolling any of the numbers 1 through 6 are all the same, making the distribution of outcomes uniform.

- **Creates a histogram** of the values in `random_numbers`.  
- **`bins=50`** → Divides the range (0 to 1) into 50 equal-width intervals. More bins = smoother histogram; fewer = chunkier.  
- **`alpha=0.75`** → Sets transparency; `1.0` is fully opaque, `0.0` is fully transparent. `0.75` makes overlapping elements more visible.  
- **`color='blue'`** → Fills histogram bars with blue.  
- **`edgecolor='black'`** → Outlines each bin in black for better visibility.  


## With Real Frequency

```python
# Generate 100,000 random numbers in [0,1)
random_numbers = np.random.rand(10)

# Plot the histogram
plt.hist(random_numbers, bins=2,  alpha=0.5, color='blue', edgecolor='white')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

# Show the plot
plt.show()
```

## With normalized frequency

```python
# Generate 100,000 random numbers in [0,1)
random_numbers = np.random.rand(100000)

# Plot the histogram
plt.hist(random_numbers, bins=50,  alpha=0.5, color='blue', edgecolor='black')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Uniformly Distributed Random Numbers')
plt.grid(True)

# Show the plot
plt.show()
```

# Normal Distribution

A **normal distribution**, also known as a **Gaussian distribution**, is a type of probability distribution that is symmetrical and bell-shaped. In this distribution:

- Most of the data points are clustered around the **mean** (average).
- As you move away from the mean, the probability of the data points decreases, but in a smooth, continuous way.
- The curve is symmetrical, meaning the left side is a mirror image of the right side.

In simpler terms, a normal distribution looks like a bell curve, where most values are near the middle, and fewer values are found as you move further away from the center. Examples include things like people's heights, test scores, or measurement errors.

```python
# Generate 10,000 random values from a normal distribution
SNormalValues = np.random.randn(10000)#mean=0, standard deviation=1

# Create a histogram
plt.hist(SNormalValues, bins=30, edgecolor='black')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generate 10,000 random values from a normal distribution with mean=100 and std=10
NormalValues = np.random.normal(0, 1, 500)

# Create a histogram
plt.hist(NormalValues, bins=35, edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=100, std=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# With and Without np.random.seed

```python
non_seeded_matrix = np.random.randint(1, 50, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)
```

```python

# With Seed
np.random.seed(10)  # Setting seed to 10
seeded_matrix = np.random.randint(1, 51, (3, 3))
print("Seeded Random Matrix:\n", seeded_matrix)

# Without Seed
np.random.seed(None)  # No seed, generates different values each time
non_seeded_matrix = np.random.randint(1, 50, (3, 3))
print("\nNon-Seeded Random Matrix:\n", non_seeded_matrix)
```
